# Exploratory data analysis

```python
ax=sns.countplot(x="Gender",data=dt1)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
ax=sns.barplot(x="Gender",y="Amount",data=dt1)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_by_gen=dt1.groupby('Gender')['Amount'].sum().sort_values(ascending=False)
print(sales_by_gen)
```

# Age

```python
Gender_wise_ag=sns.countplot(x="Age Group",data=dt1,hue="Gender")
for bars in Gender_wise_ag.containers:
    Gender_wise_ag.bar_label(bars)
plt.show()
    
```

```python
#Total amount vs Age group

sales_age=dt1.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_age
```

```python
sales_age_bar=sns.barplot(x="Age Group",y="Amount",data=sales_age)

    
```

# From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

# State

```python
dt1.columns
```

```python
# total number of orders from top 10 states

sales_state =dt1.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
```

```python
sales_state
```

```python
sns.set(rc={'figure.figsize':(20,10)})
sales_state_bar=sns.barplot(x="State",y="Orders",data=sales_state)
for bars in sales_state_bar.containers:
    sales_state_bar.bar_label(bars)
plt.show()
```

```python
dt1.columns
```

```python
# total amount of sales from top 10 states

Amount_state =dt1.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
```

```python
Amount_state
```

```python
sns.set(rc={'figure.figsize':(20,10)})
amount_state_bar=sns.barplot(x="State",y="Amount",data=Amount_state)
```
